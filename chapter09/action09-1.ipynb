### 9.1.4 模型融合

XGBoost，LightGBM两个模型融合，相加除以2

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

#### 1 数据集切分

```python
x_train, x_val, y_train, y_val = train_test_split(X_data, Y_data, test_size=0.2)
```

#### 2 构建xgb模型

```python
def build_model_xgb(x_train, y_train):
    model = xgb.XGBRegressor(n_estimator=150, learning_rate=0.1, max_depth=7)
    model.fit(x_train, y_train)
    return model
```

#### 3 构建lgb模型

```python
def build_model_lgb(x_train, y_train):
    estimator = lgb.LGBMRegressor(n_estimator=150, num_leaves=127)
    # 设置超参数的范围
    param_grid = {"learning_rate": [0.01, 0.05, 0.1, 0.2]}
    model = GridSearchCV(estimator, param_grid)
    model.fit(x_train, y_train)
    return model
```

#### 4 xgb训练

```python
%%time
model_xgb = build_model_xgb(x_train, y_train)
val_xgb = model_xgb.predict(x_val)
MAE_xgb = mean_absolute_error(y_val, val_xgb)
print("XGBoost Validation MAE: ", MAE_xgb)
```

```python
%%time
model_xgb2 = build_model_xgb(X_data, Y_data)
result_xgb = model_xgb.predict(X_test)
print("XGB预测统计结果: ")
show_stats(result_xgb)
```

#### 5 lgb训练

```python
%%time
model_lgb = build_model_lgb(x_train, y_train)
val_lgb = model_lgb.predict(x_val)
MAE_lgb = mean_absolute_error(y_val, val_lgb)
print("XGBoost Validation MAE: ", MAE_lgb)
```

```python
%%time
model_lgb2 = build_model_lgb(X_data, Y_data)
result_lgb = model_xgb.predict(X_test)
print("lgb预测统计结果: ")
show_stats(result_lgb)
```

#### 6 权重分配

```python
predict_y = (1 - MAE_lgb/(MAE_xgb + MAE_lgb))*result_lgb + (1 - MAE_xgb/(MAE_xgb + MAE_lgb))*result_xgb
```

```python
print("lgb 权重:", 1 - MAE_lgb/(MAE_xgb + MAE_lgb))
print("xgb 权重:", 1 - MAE_xgb/(MAE_xgb + MAE_lgb))
```

```python
result = pd.DataFrame()
result["SaleID"] = test_data["SaleID"]
result["price"] = predict_y
result[result["price"] < 11]["price"] = 11
result.to_csv("ans_xgb_lgb.csv", index=False)
```

```python
print("xgb&lgb 预测结果统计情况: ")
show_stats(predict_y)
```

### 9.1.5 使用神经网络

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
```

#### 1 加载数据

```python
train_data = pd.read_csv("./used_car_train_20200313.csv", sep=" ")
test_data = pd.read_csv("./used_car_testB_20200421.csv", sep=" ")
```

#### 2 处理异常值

```python
train_data["notRepairedDamage"].replace("-", "0.0", inplace=True)
train_data["notRepairedDamage"] = train_data["notRepairedDamage"].astype("float64")
```

```python
test_data["notRepairedDamage"].replace("-", "0.0", inplace=True)
test_data["notRepairedDamage"] = test_data["notRepairedDamage"].astype("float64")
```

```python
train_data.fillna(train_data.median(), inplace=True)
test_data.fillna(test_data.median(), inplace=True)
```

```python
train_data["power"].describe()
```

```python
train_data["power"][train_data["power"]>600] = 600
```

```python
train_data["power"].describe()
```

```python
test_data["power"][test_data["power"]>600] = 600
```

#### 3 特征归一化

对于神经网络而言，归一化很重要

```python
features = train_data.columns.tolist()
features.remove("price")
features.remove("SaleID")
features
```

```python
min_max_scaler = MinMaxScaler()
x = min_max_scaler.fit_transform(train_data[features].values)
x_test = min_max_scaler.transform(test_data[features].values)
```

```python
y = train_data["price"].values
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
```

#### 4 搭建模型

```python
model = keras.Sequential([
    keras.layers.Dense(300, activation="relu", input_shape=[len(features)]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(1)
])
```

```python
model.compile(loss="mae", optimizer="Adam")
```

```python
model.fit(train_x, train_y, batch_size=1024, epochs=200)
```

#### 5 模型评估

```python
print("训练集评估 MAE： ", mean_absolute_error(train_y, model.predict(train_x)))
print("测试集评估 MAE： ", mean_absolute_error(test_y, model.predict(test_x)))
```

#### 6 用全量数据训练

```python
model = keras.Sequential([
    keras.layers.Dense(300, activation="relu", input_shape=[len(features)]),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(1)
])
model.compile(loss="mae", optimizer="Adam")
model.fit(x, y, batch_size=1024, epochs=200)
```

#### 7 输出结果

```python
predict_y = model.predict(x_test)

result = pd.DataFrame()
result["SaleID"] = test_data["SaleID"]
result["price"] = predict_y
result["price"][result["price"] < 11] = 11
result.to_csv("ans_nn.csv", index=False)
```
