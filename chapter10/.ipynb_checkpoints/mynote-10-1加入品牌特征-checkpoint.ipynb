```python
import pandas as pd
import numpy as np
```

### 9.1.1 数据加载

```python
train_data = pd.read_csv("./used_car_train_20200313.csv", sep=" ")
train_data
```

### 9.1.2 数据处理

#### 1 将regDate转换为时间diff(汽车注册时间）

```python
# 方法1 减去日期最小值
min_date = np.min(train_data["regDate"])
min_date
```

```python
min_date = pd.to_datetime("19910101", format="%Y%m%d")
min_date
```

```python
train_data["regTime"] = (pd.to_datetime(train_data["regDate"], format="%Y%m%d", errors="coerce") - min_date).dt.days
train_data
```

#### 2 将createDate转换为时间diff(汽车售卖时间)

```python
np.min(train_data["creatDate"])
```

```python
train_data["creatTime"] = (pd.to_datetime(train_data["creatDate"], format="%Y%m%d", errors="coerce") - min_date).dt.days
train_data["creatTime"]
```

####  3 汽车使用时间

```python
train_data["usedTime"] = train_data["creatTime"] - train_data["regTime"]
train_data["usedTime"]
```

#### 4 测试集数据

```python
test_data = pd.read_csv("./used_car_testB_20200421.csv", sep=" ")
test_data
```

```python
test_data["regTime"] = (pd.to_datetime(test_data["regDate"], format="%Y%m%d", errors="coerce") - min_date).dt.days
test_data
```

```python
test_data["creatTime"] = (pd.to_datetime(test_data["creatDate"], format="%Y%m%d", errors="coerce") - min_date).dt.days
test_data["creatTime"]
```

```python
test_data["usedTime"] = test_data["creatDate"] - test_data["regDate"]
test_data["usedTime"]
```

#### 5 汽车品牌维度

```python
# 计算某品牌的销售统计
brand_data = train_data.groupby("brand")
brand_data.describe()
```

```python
all_info = {}
for brand_index, brand_temp in brand_data:
    info = {}
    brand_temp = brand_temp[brand_temp["price"]>0]
    info["brand_amount"] = len(brand_temp)
    info["brand_price_max"] = brand_temp.price.max()
    info["brand_price_min"] = brand_temp.price.min()
    info["brand_price_median"] = brand_temp.price.median()
    info["brand_price_sum"] = brand_temp.price.sum()
    info["brand_price_std"] = brand_temp.price.std()
    info["brand_price_mean"] = brand_temp.price.mean()
    all_info[brand_index] = info
all_info
```

```python
brand_stats = pd.DataFrame(all_info).T.reset_index().rename(columns={"index":"brand"})
brand_stats
```

```python
train_data = train_data.merge(brand_stats, how="left", on="brand")
train_data
```

```python
test_data = test_data.merge(brand_stats, how="left", on="brand")
test_data
```

### 9.1.4 模型融合

XGBoost，LightGBM两个模型融合，相加除以2

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

#### 1 数据集切分

```python
x_train, x_val, y_train, y_val = train_test_split(X_data, Y_data, test_size=0.2)
```

#### 2 构建xgb模型

```python
def build_model_xgb(x_train, y_train):
    model = xgb.XGBRegressor(n_estimator=150, learning_rate=0.1, max_depth=7)
    model.fit(x_train, y_train)
    return model
```

#### 3 构建lgb模型

```python
def build_model_lgb(x_train, y_train):
    estimator = lgb.LGBMRegressor(n_estimator=150, num_leaves=127)
    # 设置超参数的范围
    param_grid = {"learning_rate": [0.01, 0.05, 0.1, 0.2]}
    model = GridSearchCV(estimator, param_grid)
    model.fit(x_train, y_train)
    return model
```

#### 4 xgb训练

```python
%%time
model_xgb = build_model_xgb(x_train, y_train)
val_xgb = model_xgb.predict(x_val)
MAE_xgb = mean_absolute_error(y_val, val_xgb)
print("XGBoost Validation MAE: ", MAE_xgb)
```

```python
%%time
model_xgb2 = build_model_xgb(X_data, Y_data)
result_xgb = model_xgb.predict(X_test)
print("XGB预测统计结果: ")
show_stats(result_xgb)
```

```python
plt.figure(figsize=(50,20))
xgb.plot_importance(model_xgb2)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,15))
xgb.plot_importance(model_xgb2, ax=ax, height=0.5)
plt.show()
```

#### 5 lgb训练

```python
%%time
model_lgb = build_model_lgb(x_train, y_train)
val_lgb = model_lgb.predict(x_val)
MAE_lgb = mean_absolute_error(y_val, val_lgb)
print("XGBoost Validation MAE: ", MAE_lgb)
```

```python
%%time
model_lgb2 = build_model_lgb(X_data, Y_data)
result_lgb = model_lgb2.predict(X_test)
print("lgb预测统计结果: ")
show_stats(result_lgb)
```

```python
model_lgb2.best_estimator_
```

```python
fig,ax = plt.subplots(figsize=(10,10))
lgb.plot_importance(model_lgb2.best_estimator_, ax=ax, height=0.5)
plt.show()
```

#### 6 权重分配

```python
predict_y = (1 - MAE_lgb/(MAE_xgb + MAE_lgb))*result_lgb + (1 - MAE_xgb/(MAE_xgb + MAE_lgb))*result_xgb
```

```python
print("lgb 权重:", 1 - MAE_lgb/(MAE_xgb + MAE_lgb))
print("xgb 权重:", 1 - MAE_xgb/(MAE_xgb + MAE_lgb))
```

```python
result = pd.DataFrame()
result["SaleID"] = test_data["SaleID"]
result["price"] = predict_y
result[result["price"] < 11]["price"] = 11
result.to_csv("ans_xgb_lgb_add_brand__{}.csv".format(time.strftime("%Y%m%d-%H%M%S", time.localtime())), index=False)
```
