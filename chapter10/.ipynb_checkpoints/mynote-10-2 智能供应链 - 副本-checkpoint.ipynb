```python
import datetime
```

```python
# 统计最后一笔订单时间
data["order date (DateOrders)"].max()
```

```python
data["order date (DateOrders)"].value_counts()
```

```python
%%time
# 时间类型转换
data["order date (DateOrders)"] = pd.to_datetime(data["order date (DateOrders)"])
# 统计最后一笔订单时间
data["order date (DateOrders)"].max()
```

```python
present = datetime.datetime(2018, 2, 1)
present
```

```python
data.columns
```

```python
customer_seg = data.groupby("Order Customer Id").agg({"order date (DateOrders)": lambda x: (present-x.max()).days,\
                                                     "Order Id" : lambda x: len(x), "Sales" : lambda x: x.sum()})
customer_seg
```

```python
quantiles = customer_seg.quantile(q=[0.25, 0.5, 0.75])
quantiles
```

```python
# R_Value越小，则R_Score越大
def R_Score(a, b, c):
    if a <= c[b][0.25]:
        return 4
    if a <= c[b][0.5]:
        return 3
    if a <= c[b][0.75]:
        return 2
    return 1
```

```python
# Frequency，Money越大，则FM_Score越大
def FM_Score(a, b, c):
    if a <= c[b][0.25]:
        return 1
    if a <= c[b][0.5]:
        return 2
    if a <= c[b][0.75]:
        return 3
    return 4
```

```python
customer_seg.rename(columns={"order date (DateOrders)" : "R_Value",
                               "Order Id" : "F_Value",
                               "Sales" : "M_Value"}, inplace=True)
customer_seg
```

```python
quantiles.rename(columns={"order date (DateOrders)" : "R_Value",
                               "Order Id" : "F_Value",
                               "Sales" : "M_Value"}, inplace=True)
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
customer_seg["R_Score"] = customer_seg["R_Value"].apply(R_Score, args=("R_Value", quantiles))
customer_seg["F_Score"] = customer_seg["F_Value"].apply(FM_Score, args=("F_Value", quantiles))
customer_seg["M_Score"] = customer_seg["M_Value"].apply(FM_Score, args=("M_Value", quantiles))
```

```python
 customer_seg
```

```python
def RFM_User(df):
    if df["M_Score"] > 2 and df["F_Score"] > 2 and df["R_Score"] > 2:
        return "重要价值用户"
    if df["M_Score"] > 2 and df["F_Score"] <= 2 and df["R_Score"] > 2:
        return "重要发展用户"
    if df["M_Score"] > 2 and df["F_Score"] > 2 and df["R_Score"] <= 2:
        return "重要保持用户"
    if df["M_Score"] > 2 and df["F_Score"] <= 2 and df["R_Score"] <= 2:
        return "重要挽留用户"
    
    if df["M_Score"] <= 2 and df["F_Score"] > 2 and df["R_Score"] > 2:
        return "一般价值用户"
    if df["M_Score"] <= 2 and df["F_Score"] <= 2 and df["R_Score"] > 2:
        return "一般发展用户"
    if df["M_Score"] <= 2 and df["F_Score"] > 2 and df["R_Score"] <= 2:
        return "一般保持用户"
    if df["M_Score"] <= 2 and df["F_Score"] <= 2 and df["R_Score"] <= 2:
        return "一般挽留用户"
```

```python
customer_seg["Customer_Segmentation"] = customer_seg.apply(RFM_User, axis=1)
customer_seg
```
